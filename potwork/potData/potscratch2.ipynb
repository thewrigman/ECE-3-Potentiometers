```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
POTDATA = pd.read_csv('pot1/pot1.csv')
X = POTDATA['median'].values.reshape(-1,1)
y = POTDATA['mm'].values
coeffs = np.polyfit(X.flatten(), y, 4)
fitted_function = np.poly1d(coeffs)
plt.figure(figsize=(10,10))
plt.plot(X,y, 'o', label='True Data')
plt.plot(X, fitted_function(X), label='Fitted Polynomial')
print(coeffs)
```
